```python
import os
import re
import numpy as np
import pandas as pd
import h5py as h5
import matplotlib.pyplot as plt
```

```python
input_path = '/scratch/sdodl001/ENCODE/eval/GM12878/'
output_path = '/scratch/sdodl001/ENCODE/eval/GM12878/results_aggregated/'
```

```python
folders = os.listdir(input_path)
folders.sort()
```

```python
def regex_match(start_with, end_with, folder):
    m = re.search(start_with+end_with, folder)
    if m!=None:
        return True
    else:
        return False
```

```python
### Aggregating results from single strand evaluation reports
df_ss = pd.DataFrame()
for folder in folders:
    if regex_match('^dna.', '*_ss_report$', folder):
        files = os.listdir(input_path+folder)
        for file in files:
            match = re.match('^report', file)
            if match!=None:
                #print('file: ', file)
                file_path = input_path+folder+'/'+file
                report_data = pd.read_csv(file_path, sep='\t')
                report_data = report_data.loc[:, ['value']]
                if len(report_data) == 7:
                    report_data = report_data.T
                    report_data.insert(loc=0, column='name', value=[folder])
                    report_data.reset_index(drop=True, inplace=True)
                    df_ss = pd.concat([df_ss, report_data])
                    
df_ss.columns = ['data_name', 'acc', 'auc', 'f1', 'mcc', 'n', 'tnr', 'tpr']
file_name = 'GM12878_ss_results_aggregated.csv'
df_ss.to_csv(output_path+file_name, index=False)
print('aggregared results are saved to :', output_path)
print('Done !')
```
